```python
import re

# List of profession-related keywords
professions = [
    'doctor',
    'engineer',
    'teacher',
    'lawyer',
    'chef',
    'artist',
    # Add more profession keywords as needed
]

# Example sentence
sentence = "John, the doctor, treated patients and discussed cases with his colleagues."

# Perform keyword matching
matched_professions = []
for profession in professions:
    # Create a regular expression pattern to match the profession keyword
    pattern = r'\b{}\b'.format(profession)
    # Use regex to find matches in the sentence
    matches = re.findall(pattern, sentence, flags=re.IGNORECASE)
    # If a match is found, add the profession to the list of matched professions
    if matches:
        matched_professions.extend(matches)

# Print the matched professions
print(matched_professions)
```
